```python
import os
import pandas as pd
```

```python
arquivos = os.listdir("./livros_filosofia")  # Lista os arquivos no diret√≥rio atual

dict_df = {"autor": [], "titulo": [], "texto": []}

for i, arquivo in enumerate(arquivos):
    with open("./livros_filosofia/" + arquivo, "r") as f:

        texto = f.read()
        autor, titulo = arquivo.split("-")[0], arquivo.split("-")[1]

        titulo = titulo.split(".txt")[0]
        print(i, autor, titulo)

        texto_split = texto.split("*** START")[1]
        texto_split = texto_split.split("*** END")[0]
        
        texto = texto_split[30::]
        
        dict_df["autor"].append(autor)
        dict_df["titulo"].append(titulo)
        dict_df["texto"].append(texto)

df = pd.DataFrame(dict_df)

df.head()
```

```python
df.to_csv("./dados/livros_filosofia.csv", index=False)
```
