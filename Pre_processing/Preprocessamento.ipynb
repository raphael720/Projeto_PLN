```python
import streamlit as st
import nltk
import pandas as pd
import re
from textblob import TextBlob
```

### Leitura dos dados

```python
df = pd.read_csv('./dados/livros_filosofia.csv')
df
```

### Tratamento dos dados

```python


autores_contagem = df["autor"].value_counts()

print(autores_contagem[:5])

autores_selecionados = autores_contagem[:5].index.tolist()

df = df[df["autor"].isin(autores_selecionados)]

# print(df["autor"].value_counts())

print(len(df.autor.unique()))
print(df.autor.unique())
```

```python
import re

def clean(text):
    text = str(text).lower()
    text = re.sub('\[*\]', '', text)
    text = re.sub('\*+', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<*>+', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'\b\w{1,2}\b', '', text)
    return text

df["texto"] = df["texto"].apply(clean)
```

```python
def sentences_tokenization(text):
    sentences = []
    blob = TextBlob(text)
    for sentence in blob.sentences:
        sentences.append(str(sentence))
    
    return sentences

df['sentences'] = df['texto'].apply(sentences_tokenization)
```

### Pré-processamento

```python
# criar uma lista de sentenças para todos os autores
all_sentences = []
all_authors = []


for sentences, autor in df[['sentences', 'autor']].values.tolist():
    all_sentences.extend(sentences)
    all_authors.extend([autor] * len(sentences))


# criar um dataset novo com sentença e autor
new_df = pd.DataFrame({'sentences': all_sentences, 'autor': all_authors})

print(new_df.shape)
print(new_df.head())
```

```python
new_df.to_csv('./dados/sentencas_autores.csv', index=False)
```
